```python
%%time

reads_col = 'unique_junction_reads'

se_dfs = []

def get_junction_reads(junction, sj_samples_grouped, reads_col):
    try:
        return sj_samples_grouped.get_group(junction)[reads_col].astype(float)
    except KeyError:
        return pd.Series()

for junctions, s in exons_to_junctions.groupby(exons_to_junctions):
    junctions = eval(junctions)
    
    # "1", "2", and "3" in this case are in genomic order, not in transcript order
    exon12 = get_junction_reads(junctions[0], sj_samples_grouped, reads_col)
    exon23 = get_junction_reads(junctions[1], sj_samples_grouped, reads_col)
    exon13 = get_junction_reads(junctions[2], sj_samples_grouped, reads_col)
    
    df = pd.concat([exon12, exon23, exon13], axis=1, keys=['exon12', 'exon23', 'exon13'])
#     df.columns = 
    df = df.fillna(0)
    
    df['$\Psi$'] = (df.exon12 + df.exon23).divide(df.exon12 + df.exon23 + 2*df.exon13)
    df['junction1'], df['junction2'], df['junction3'] = junctions
    df = df.reset_index()
    df = df.rename(columns={'index': 'sample_id'})
#     print df.head()
    se_dfs.append(df)
se_df = pd.concat(se_dfs, ignore_index=True)
```

For samples that share a donor (acceptor), add the acceptor (donor) to other samples, with zero reads. This ***explicitly*** adds all possible `donor, acceptor` pairs to each sample, so the other samples will properly get $\Psi$ scores of 1 for the `donor, acceptor` pairs they have not seen.

    Sample 1:  
    chr1:100-200: 25 reads
    [   ]-------------[   ]

    [   ]----------------------[    ]
    

**To:**

    Sample 1:  
    chr1:100-200: 25 reads
    [   ]-------------[   ]
    chr1:100:250: 0 reads
    [   ]----------------------[    ]
    Sample 2:  
    chr1:100-200: 0 reads
    [   ]-------------[   ]
    chr1:100:250: 40 reads

### Calculate Percent spliced-in/Psi/$\Psi$ from splice junctions

This method of getting psi scores is 5x faster than using `lambda x: x/x.sum()`

```python
%%time

splice_sites = {r'$\Psi_5$': 'intron_start', r'$\Psi_3$': 'intron_stop'}

for psi, ss in splice_sites.items():
    sj_appended = sj_appended.sort(full_index)
    groupby = ['sample_id', 'chrom', ss, 'strand']
    reads = sj_appended.groupby(groupby)[reads_col].sum()
    reads.name = reads_col
    
    # sort_index() is necessary for the rows of sj_appended
    # to be exactly the right order and match with psi_scores
    sj_appended = sj_appended.set_index(groupby).sort_index()
    psi_scores = sj_appended[reads_col].divide(reads).fillna(0)
    sj_appended[psi] = psi_scores.values
    sj_appended = sj_appended.reset_index()
```
